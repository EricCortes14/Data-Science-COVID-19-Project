### Import Libraries

```python
#import libraries
import pandas as pd
```

### Data Set Import

```python
#import data sets and save them as dataframes using pandas
housing = pd.read_csv("../data/ACSDP1Y2019.DP04_data_with_overlays_2020-09-17T112419.csv")
main_dataset = pd.read_csv("../data/covid_large.csv")
```

### Data Set Manipulation

```python
#remove unused columns
housing_2 = housing[['NAME', 'DP04_0046E', 'DP04_0047E', 'DP04_0048E', 'DP04_0049E']]

#remove unused row to facilitate merge
housing_3 = housing_2.drop([0])

#remane column names
housing_3.rename(columns = {'NAME':'County Name'}, inplace = True)
housing_3.rename(columns = {'DP04_0046E':'Owner-Occupied Unit'}, inplace = True)
housing_3.rename(columns = {'DP04_0047E':'Renter-Occupied Unit'}, inplace = True)
housing_3.rename(columns = {'DP04_0048E':'Average Household Size of Owner-Occupied Unit'}, inplace = True)
housing_3.rename(columns = {'DP04_0049E':'Average Household Size of Renter-Occupied Unit'}, inplace = True)

#make country name lowercase and remove state name after comma
housing_3['County Name'] = housing_3['County Name'].apply(lambda cn: cn.lower().strip())
housing_3['County Name'] = housing_3['County Name'].apply(lambda cn: cn.split(",")[0])
```

### Merge Data Sets and Display

```python
#merge covid-19 data set and housing data set
new_dataset = pd.merge(main_dataset, housing_3)

#display merged data set
new_dataset
```
