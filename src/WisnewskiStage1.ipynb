```python
# formats data
```

```python
#converts datatype to datetime format
covid_data['Date'] = covid_data.Date.apply(lambda d: datetime.strptime(d, '%Y-%m-%d'))
```

```python
# Selecting State Data for Virginia for the Week of Sept 7
```

```python
#Select data by state
covid_VAdata=covid_data.loc[covid_data['State'] == 'VA']
#mask to select state data by date
start_date=datetime(2020,9,6)
end_date=datetime(2020,9,15)
mask = (covid_VAdata['Date'] > start_date) & (covid_VAdata['Date'] <= end_date)
#apply mask
virginia = covid_VAdata.loc[mask]
virginia
```

```python
# Summing County Cases and Deaths per day
```

```python
virginia.set_index('Date')
cases = virginia.groupby('Date')['Num of Cases'].sum()
deaths = virginia.groupby('Date')['Num of Deaths'].sum()
```

```python
# VA COVID-19 Cases Week of Sept 7
```

```python
ax=cases.plot(title='Number of COVID-19 Cases',figsize=(10,5), grid=True,legend=True)
ax.set_ylabel("COVID-19 cases")
```

```python
## COVID-19 cases in Virginia have been increasing this past week steadily
```

```python
#VA COVID-19 Deaths Week of Sept 7 steadily
```

```python
ax=deaths.plot(title='Number of COVID-19 Deaths',figsize=(10,5), grid=True,legend=True)
ax.set_ylabel("COVID-19 Deaths")
```
