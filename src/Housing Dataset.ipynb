# Housing Dataset

### Import Libraries

```python
#import libraries
import pandas as pd
```

### Import Dataset

```python
#import datasets and save them as dataframes using pandas
housing = pd.read_csv("../data/ACSDP1Y2019.DP04_data_with_overlays_2020-09-17T112419.csv")
main_dataset = pd.read_csv("../data/covid_large.csv")
```

### Dataset Manipulation

```python
#display original dataset
housing.head()
```

```python
#remove unused columns
housing = housing[['NAME', 'DP04_0046E', 'DP04_0047E', 'DP04_0048E', 'DP04_0049E', 'DP04_0058E']]
housing
```

```python
#remove unused row to facilitate merge
housing = housing.drop([0])
housing.head()
```

```python
#remane column names
housing.rename(columns = {'NAME':'County Name'}, inplace = True)
housing.rename(columns = {'DP04_0046E':'Owner-Occupied Unit'}, inplace = True)
housing.rename(columns = {'DP04_0047E':'Renter-Occupied Unit'}, inplace = True)
housing.rename(columns = {'DP04_0048E':'Average Household Size of Owner-Occupied Unit'}, inplace = True)
housing.rename(columns = {'DP04_0049E':'Average Household Size of Renter-Occupied Unit'}, inplace = True)
housing.rename(columns = {'DP04_0058E':'No Vehicles Available'}, inplace = True)
housing.head()
```

```python
#make country name lowercase and remove state name after comma
housing['County Name'] = housing['County Name'].apply(lambda cn: cn.lower().strip())
housing['County Name'] = housing['County Name'].apply(lambda cn: cn.split(",")[0])
housing.head()
```

### Merge Datasets

```python
#merge covid-19 dataset and housing dataset
merged_dataset = pd.merge(main_dataset, housing)

#display merged dataset
merged_dataset
```

## Exporting to csv

```python
#merged dataset is exported as csv into data file
merged_dataset.to_csv('../data/housing_merge.csv',index=False)
```
