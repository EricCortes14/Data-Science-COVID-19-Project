#### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Read data from CSV Files - covid_large.csv and Hospital_beds.csv

```python
# read data from covid_large csv file
covid_large_df = pd.read_csv("../data/covid_large.csv")
covid_large_df.head()
```

```python
# read data from Hospital_beds csv file
hospital_beds_df = pd.read_csv("../data/Definitive_Healthcare%3A_USA_Hospital_Beds.csv")
hospital_beds_df.head()
```

### Calculation of COVID-19 data trends for last week of the data for CALIFORNIA State

```python
# Extracting California State Data from covid_large_df between 09/07/2020 to 09/14/2020
COVID_CA = covid_large_df.loc[(covid_large_df["countyFIPS"] != 0) & (covid_large_df["State"] == "CA") & (covid_large_df["Date"] >= "2020-09-07")]
COVID_CA.head()
```

#### Statistics of COVID_CA dataset

```python
#Statistics of COVID_CA variables
COVID_CA.describe()
```

<li>Maximum Number of Confirmed Cases for a specific day across week - 254656</li>
<li>Maximum Number of Deaths for a specific day across week - 6231</li>
<li>Minimum Number of Confirmed Cases for a specific day across week - 2</li>
<li>Minimum Number of Deaths for a specific day across week - 0 (Some counties in CA didnt report any deaths)</li>

#### Overall Trend of "Num of Cases" and "Num of Deaths" for all counties each day in California for the week 09/07/2020 to 09/14/2020

```python
# Aggregation of all the counties data by grouping on Date 
COVID_CA_Agg = COVID_CA.groupby('Date', as_index=False).sum()
# Dropping 'countyFIPS' column 
COVID_CA_Agg = COVID_CA_Agg.drop(['countyFIPS'], axis=1)
COVID_CA_Agg
```

```python
# Normalizing "Num of Cases" and "Num of Deaths" Data
normalize = lambda x: (x - x.mean())/x.std()
COVID_CA_Agg["Num of Cases"] = COVID_CA_Agg["Num of Cases"].transform(normalize)
COVID_CA_Agg["Num of Deaths"] = COVID_CA_Agg["Num of Deaths"].transform(normalize)
```

```python
# Plot displaying "Date" as x-axis and normalized "Num of Cases" and "Num of Deaths" Data on y-axis
plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(COVID_CA_Agg["Date"], COVID_CA_Agg["Num of Cases"], label='Number of Confirmed Cases', marker='o')
plt.plot(COVID_CA_Agg["Date"], COVID_CA_Agg["Num of Deaths"], label='Number of Deaths', marker='x')
plt.legend(loc="upper left")
```

<li>From the plot we can determine that the overall trend for all counties in California for the week 09/07/2020 to 09/14/2020 is <b>increasing</b>.</li>
<li> We can also observe that the "Num of Deaths" is less compared to "Num of Confirmed Cases" from 09/07/2020 to 09/09/2020. However, there is shift in trend of the "Num of Deaths" after 09/09/2020 where it is increased compared to "Num of Cases" </li>
<li> The "Num of Deaths" is almost equal to "Num of Confirmed Cases" between 09/09/2020 and 09/10/2020</li>


#### County with Maximum Num of Cases and Num of Deaths in CA between 09/07/2020 and 09/14/2020

```python
#Using GroupBy countyFIPS on COVID_CA to get max values of all the columns for each county
COVID_CA_ = COVID_CA.groupby('countyFIPS', as_index=False).max()
print(COVID_CA_.loc[COVID_CA_['Num of Cases'].idxmax() & COVID_CA_['Num of Deaths'].idxmax()])
```

<p> From the "Date" variable we can say that their is increase in COVID through the week as the highest Number of COVID Cases and Deaths are found in Los Angeles County at the end of week 14TH September,2020.  </p>

#### County with Minimum Num of Cases and Num of Deaths in CA between 09/07/2020 and 09/14/2020

```python
#Using GroupBy countyFIPS on COVID_CA to get min values of all the columns for each county
COVID_CA_ = COVID_CA.groupby('countyFIPS', as_index=False).min()
print(COVID_CA_.loc[COVID_CA_['Num of Cases'].idxmin()])
```

<p> The minimum Number of COVID Cases are found in alpine County at the start of week 7TH September,2020.  </p>

### Merge of COVID-19 dataset with Hospital Beds Enrichment dataset 

#### Preprocessing of Hospital Beds Enrichment dataset

```python
#Droppping few columns which might be repeatitive after merge
hospital_beds_df=hospital_beds_df.drop(['X','Y','HQ_ADDRESS1','HQ_STATE','COUNTY_NAME','STATE_NAME','STATE_FIPS','CNTY_FIPS'], axis=1)

#Dropping rows having NAN in FIPS column
hospital_beds_df = hospital_beds_df[hospital_beds_df['FIPS'].notna()]

#Converted all NAN cells to 0
hospital_beds_df = hospital_beds_df.fillna(0)

#Converted few float columns to int
hospital_beds_df = hospital_beds_df.astype({'FIPS': 'int','NUM_LICENSED_BEDS':'int','NUM_STAFFED_BEDS':'int','PEDI_ICU_BEDS':'int'})

hospital_beds_df.head()
```

#### Merge of covid_large_df and hospital_beds_df

```python
#Merged covid_large_df and hospital_beds_df using inner join 
#by joining on "countyFIPS" in covid_large_df with "FIPS" in hospital_beds_df
COVID_HospitalBeds_merged_df = pd.merge(covid_large_df,hospital_beds_df,how="inner",left_on="countyFIPS",right_on="FIPS")
COVID_HospitalBeds_merged_df = COVID_HospitalBeds_merged_df.drop(['OBJECTID','HQ_ADDRESS','FIPS'], axis=1)
COVID_HospitalBeds_merged_df
```

#### Exporting COVID_HospitalBeds_merged_df to path '../data/COVID_HospitalBeds_merged_df.csv'

```python
COVID_HospitalBeds_merged_df.to_csv('../data/COVID_HospitalBeds_merged_df.csv',index=False)
```
